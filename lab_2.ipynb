## Filter design

```python
# TODO: Design the filters
def bandpass_filter(x, lowcut = 3750, highcut = 3850, fs = fs, order=4):
    nyq = fs / 2                     # Nyquist-frekvens
    low = lowcut / nyq
    high = highcut / nyq

    b, a = butter(order, [low, high], btype='band')
    y = filtfilt(b, a, x)            # Zero-phase filtering
    return y


```
